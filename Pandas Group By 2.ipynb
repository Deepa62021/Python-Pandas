```python
import pandas as pd
import numpy as np
```

```python
dfsales = pd.read_csv("gbsales.csv")
dfsales
```

```python
dfsales.drop("Status", axis = 1, inplace = True)
dfsales
```

```python
# Sales under each Manager

gbmang = dfsales.groupby("Manager")
gbmang
```

```python
for x , y in gbmang:
    print(x)
    print()
    print(y)
    print("*****************************")
    print()
```

```python
#  there are two  managers :
#  Debra Henley and Fred Anderson 

dfdebra = gbmang.get_group("Debra Henley")
dfdebra
```

```python
# Applying the function

dfdebra["Price"].max()
```

```python
dfdebra[dfdebra["Price"] == dfdebra["Price"].max()]
```

```python
# Similarly we can perform the steps under the Manager "Fred Anderson"

dffred = gbmang.get_group("Fred Anderson")
dffred
```

```python
dffred["Price"].min()
```

```python
dffred[dffred["Price"] == dffred["Price"].min()]
```
