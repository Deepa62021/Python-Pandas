## EDA using Pandas Group By

#### Group by is applicable in three steps process:
* Splitting the data
* Applying conditin/function
* Combining the results

```python
## This comes under feature engineering
```

```python
import pandas as pd
import numpy as np
```

```python
dfgen = pd.read_csv("Gb_gender.csv")
dfgen
```

```python
# Finding the tallest people in the male and female groups
```

```python
# Split the data using "Gender"
dfgen["Gender"].unique()
```

```python
gbgen = dfgen.groupby("Gender")
gbgen
```

```python
gbgen.groups
```

```python
dfgen.head()
```

```python
gbgen.head()
```

```python
for x , y in gbgen:
    print(x)
    print()
    print(y)
    print()
    print("**************************")
    print()
```

```python
# Finding the tallest male and tallest female
```

```python
dffemale = gbgen.get_group("Female")
dffemale
```

```python
dffemale.shape
```

```python
# Applying function

dffemale["Height"].max()
```

```python
# how many people having 179 height?
dffemale["Height"] == dffemale["Height"].max()
```

```python
# Result

fres = dffemale[dffemale["Height"] == dffemale["Height"].max()]
fres
```

```python
dfmale = gbgen.get_group("Male")
dfmale
```

```python
# Applying function

dfmale["Height"].max()
```

```python
# Result

mres = dfmale[dfmale["Height"] == dfmale["Height"].max()]
mres
```

```python
# Combining the results

results = pd.concat([fres, mres])
results
```

```python
# Similarly we can split the data using two columns

gbgen2 = dfgen.groupby(["Gender", "City"])
gbgen2
```

```python
for x , y in gbgen2:
    print(x)
    print()
    print(y)
    print("********************************")
    print()
    
    
```

```python
gbgen2.get_group(("Female" , "Hyderabad"))
```
